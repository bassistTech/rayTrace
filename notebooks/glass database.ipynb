Glass data is from this site:

M. N. Polyanskiy, "Refractive index database," https://refractiveindex.info. Accessed on 2023-11-07.

For now, I'm using the refractive index formula from the Schott catalog, which is restricted to visible wavelengths. Really, it's just to try out whether my program can handle glasses at all.

```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import opticsRayTrace.rayTraceTools as rtt
```

```python
model = rtt.load_glass('N-BK7')
model
```

```python
mm = np.linspace(0.0003, 0.0025)
def glass_index(model, mm):
    microns = mm*1000
    b1, c1, b2, c2, b3, c3 = model['coeffs']
    l2 = microns**2
    return np.sqrt(1 + b1*l2/(l2 - c1) + b2*l2/(l2 - c2) + b3*l2/(l2 - c3))

n = glass_index(model, mm)
plt.plot(microns*1e6, n)
plt.show()
```
